```python
import numpy as np
import tensorflow as tf
```

## 1. tensorflow way...

```python
x_data = [[73., 80., 75.],
          [93., 88., 93.],
          [89., 91., 90.],
          [96., 98., 100.],
          [73., 66., 70.]]
y_data = [[152.],
          [185.],
          [180.],
          [196.],
          [142.]]
```

```python
W = tf.Variable(tf.random.normal([3, 1], name="weight"))
b = tf.Variable(tf.random.normal([1], name="bias"))
```

```python
@tf.function
def hypothesis(X):
    return tf.matmul(X, W) + b
```

```python
@tf.function
def mse_loss(y_pred, y):
    return tf.reduce_mean(tf.square(y_pred - y))
```

```python
optimizer = tf.optimizers.SGD(learning_rate=1e-5)
```

```python
epochs = 2000 #(or however many iterations you want it to run)

for step in range(epochs):
    with tf.GradientTape() as tape:
        
        y_pred = hypothesis(x_data)   
        cost = mse_loss(y_pred, y_data)
        
    gradients = tape.gradient(cost, [W, b])

    optimizer.apply_gradients(zip(gradients, [W, b]))
    
    if step % 10 == 0:
        print(step, W.numpy(), b.numpy(), cost.numpy())
```

```python
tf.cast(W, dtype='float32')
tf.cast(b, dtype='float32')
```

```python
y_predict = hypothesis([[72., 93., 90.]])
print(y_predict)
```

## 2. keras way...

```python
tf.model = tf.keras.Sequential()
```

```python
# input_dim=3 gives multi-variable regression
tf.model.add(tf.keras.layers.Dense(units=1, input_dim=3))
tf.model.add(tf.keras.layers.Activation("linear")) # this line can be omitted, as linear activation is default
```

```python
sgd=tf.keras.optimizers.SGD(learning_rate=1e-5)
```

```python
tf.model.compile(loss="mse", optimizer=sgd, metrics=["mse"])
tf.model.summary()
```

```python
history = tf.model.fit(x_data, y_data, epochs=500)
```

```python
y_predict = tf.model.predict(np.array([[72., 93., 09.]]))
print(y_predict)
```
